```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
data = pd.read_csv('/content/householdtask3.csv')
```

```python
data.head()
```

```python
#Scatter plot

plt.scatter(data['year'], data['own'])
plt.title("Scatter Plot")

plt.xlabel('year')
plt.ylabel('own')

plt.show()
```

```python
#line_chart

plt.plot(data['year'])
plt.plot(data['own'])

plt.title("Line chart with Year against Own")

plt.xlabel('year')
plt.ylabel('own')

plt.show()
```

```python
plt.bar(data['year'], data['own'])
plt.title("Bar chart")

plt.xlabel('year')
plt.ylabel('own')

plt.show()
```

```python
plt.plot(data['year'])
plt.plot(data['income'])

plt.title("Line chart")

plt.xlabel('year')
plt.ylabel('income')

plt.plot(data['year'], label = 'Y')
plt.plot(data['income'], label = 'I')

plt.legend(fontsize = 'x-large', loc = 'upper left')

plt.show()
```

```python
plt.bar(data['year'], data['income'])
plt.title("Bar chart")

plt.xlabel('year')
plt.ylabel('income')


plt.show()
```

```python
plt.plot(data['year'])
plt.plot(data['expenditure'])

plt.title("Line_Chart")

plt.xlabel('year')
plt.ylabel('expenditure')

plt.plot(data['year'], label = "Y")
plt.plot(data['expenditure'], label = "E")

plt.legend(labelcolor = 'Pink', facecolor = 'blue',
           edgecolor = 'Brown', fontsize = 'xx-large')
```

```python
plt.plot(data['age'])
plt.plot(data['income'])

plt.title("Line chart with Age against Income")

plt.xlabel('age')
plt.ylabel('income')

plt.plot(data['age'], label='A')
plt.plot(data['income'], label = 'I')

plt.legend(fontsize = 'x-large', loc = 'upper left')

plt.show()
```

```python
plt.bar(data['age'], data['income'])
plt.title("Age v/s Income")

plt.xlabel('age')
plt.ylabel('income')


plt.show()
```
